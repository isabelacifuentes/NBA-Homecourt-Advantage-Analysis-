```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv("nba_data/game.csv", parse_dates=["game_date"])

# Quick check
print(df.columns)
print(df.head())
print(df.info())
```

```python
#Summary of points scored, rebounds, assists...
df[['pts_home', 'pts_away', 'reb_home', 'reb_away', 'ast_home', 'ast_away']].describe()
```

```python
#Summary of points scored, rebounds, assists...
df[['pts_home', 'pts_away', 'reb_home', 'reb_away', 'ast_home', 'ast_away']].describe()
```

```python
#Win Rate by Season; Track how the home win percentage changes over time.
home_win_rate_by_season = df.groupby('season_id')['wl_home'].apply(lambda x: (x == 'W').mean())
home_win_rate_by_season.plot(marker='o')
plt.title("Home Win Rate by Season")
plt.ylabel("Win %")
plt.show()
```

```python
#Point differential distribution
#Spread of outcomes; shows how often games are blowouts vs close

sns.histplot(df['pts_home'] - df['pts_away'], bins=30, kde=True, color='skyblue', label='Point Differential')
plt.axvline(0, color='red', linestyle='--', label='No Advantage (0)')
plt.title("Distribution of Home Point Differential")
plt.xlabel("Home - Away Points")
plt.ylabel("Number of Games")
plt.legend()  # <-- adds a legend/key
plt.show()

```

```python
#Win rate at home; top 10 best performing teams at home
team_home_win = df.groupby('team_abbreviation_home')['wl_home'].apply(lambda x: (x == 'W').mean()).sort_values(ascending=False)
team_home_win.head(10)  # top 10 strongest home teams
```

```python
# Calculate home vs away win %
home_win_rate = (df['wl_home'] == 'W').mean()
away_win_rate = (df['wl_away'] == 'W').mean()
print(f"Home Win %: {home_win_rate:.2f}")
print(f"Away Win %: {away_win_rate:.2f}")

# Visualize home vs away win %
win_df = pd.DataFrame({
    'home_away': ['Home', 'Away'],
    'win_rate': [home_win_rate, away_win_rate]
})

sns.barplot(data=win_df, x='home_away', y='win_rate')
plt.title("Win Percentage: Home vs Away")
plt.ylabel("Win %")
plt.show()

# Point differential: home minus away
df['point_diff'] = df['pts_home'] - df['pts_away']
sns.boxplot(data=df, y='point_diff')
plt.title("Point Differential (Home - Away)")
plt.ylabel("Points")
plt.show()

#On the box plot, the positive values scored more than the away team
#Negative values mean the away team scored more than home team
```

```python
# Compare raw points scored by home vs away teams
pts_df = pd.melt(df, value_vars=['pts_home', 'pts_away'],
                 var_name='Game_Type', value_name='Points')

# Clean labels
pts_df['Game_Type'] = pts_df['Game_Type'].map({'pts_home':'Home', 'pts_away':'Away'})

# Boxplot comparison
sns.boxplot(data=pts_df, x='Game_Type', y='Points', palette=['skyblue','lightcoral'])
plt.title("Points Scored: Home vs Away Teams")
plt.ylabel("Points Scored")
plt.xlabel("")
plt.show()
```

```python
# Sort by team and date for home games
df_home = df.sort_values(['team_name_home', 'game_date'])
df_home['prev_game_date'] = df_home.groupby('team_name_home')['game_date'].shift(1)
df_home['days_rest'] = (df_home['game_date'] - df_home['prev_game_date']).dt.days

sns.histplot(df_home['days_rest'].dropna(), bins=20)
plt.title("Distribution of Days Rest Between Home Games")
plt.xlabel("Days Rest")
plt.ylabel("Number of Games")
plt.show()
```

```python
# Home win % by season
df['season_year'] = df['season_id'].apply(lambda x: int(str(x)[:4]))
season_trend = df.groupby(['season_year'])['wl_home'].apply(lambda x: (x=='W').mean()).reset_index()

sns.lineplot(data=season_trend, x='season_year', y='wl_home', marker='o')
plt.title("Home Win % Over Seasons")
plt.xlabel("Season")
plt.ylabel("Win %")
plt.xticks(rotation=45)
plt.show()
```

```python
#Statistcal Testing
import statsmodels.api as sm

# Proportion of home wins
home_win_rate = (df['wl_home'] == 'W').mean()
n_games = len(df)

# Null hypothesis: home win probability = 0.5
stat, pval = sm.stats.proportions_ztest(
    count=(df['wl_home'] == 'W').sum(),
    nobs=n_games,
    value=0.5
)

print(f"Home Win Rate: {home_win_rate:.3f}")
print(f"Z-test p-value: {pval:.5f}")
```

```python
#P_Value is less than 0.05 so reject null; home win rate is significantly different from 50%
```

```python
#Do home teams score more points than away teams?
from scipy import stats

tstat, pval = stats.ttest_rel(df['pts_home'], df['pts_away'])
print(f"T-statistic: {tstat:.3f}, p-value: {pval:.5f}")
```

```python
#P-value is less than 0.05 so we reject null; home teams score significantly more than away teams
```
